#### 1. Load and Inspect

```python
import pandas as pd
import numpy as np

df = pd.read_csv('../data/raw/procurement_raw.csv', 
                 parse_dates=['PO_Date', 'Expected_Delivery', 'Actual_Delivery'])

print(f"Dataset Shape: {df.shape}")
print(df.info())
```

#### 2. Detecting "The Impossible" (Negative Lead Times)

```python
df['Actual_Lead_Time'] = (df['Actual_Delivery'] - df['PO_Date']).dt.days

impossible_records = df[df['Actual_Lead_Time'] < 0]

print(f"ALERT: Found {len(impossible_records)} records with negative lead times.")
print(impossible_records[['PO_Number', 'Supplier', 'PO_Date', 'Actual_Delivery']])

df.loc[df['Actual_Lead_Time'] < 0, 'Actual_Delivery'] = df['Expected_Delivery']
df['Actual_Lead_Time'] = (df['Actual_Delivery'] - df['PO_Date']).dt.days
```

#### 3. The "Price Creep" Audit (OEM Variance)

```python
df['Price_Diff'] = df['Unit_Price_Final'] - df['Unit_Price_PO']
df['Price_Variance_Pct'] = (df['Price_Diff'] / df['Unit_Price_PO']) * 100

variance_report = df.groupby('Category')['Price_Variance_Pct'].mean().reset_index()
print("Average Price Variance by Category:")
print(variance_report)

df['High_Price_Risk'] = df['Price_Variance_Pct'] > 5
```

#### 4. Logistics Reliability (OTD % Analysis)

```python
df['Days_Late'] = (df['Actual_Delivery'] - df['Expected_Delivery']).dt.days

def get_otd_status(days):
    if days <= 0: return 'On-Time'
    if days <= 7: return 'Minor Delay'
    return 'Critical Delay'

df['OTD_Status'] = df['Days_Late'].apply(get_otd_status)

print("Logistics Status Distribution:")
print(df['OTD_Status'].value_counts(normalize=True) * 100)
```

#### 5. Quantity Accuracy (The "Short Shipment" Audit)

```python
df['Qty_Accuracy_Rate'] = (df['Received_Qty'] / df['Ordered_Qty']) * 100
df['Is_Short_Shipment'] = df['Received_Qty'] < df['Ordered_Qty']

short_summary = df.groupby('Supplier')['Is_Short_Shipment'].sum().sort_values(ascending=False)
print("Top 5 Suppliers with Short Shipments:")
print(short_summary.head(5))
```

#### 6. Export Processed Data

```python
import os
os.makedirs('../data/processed', exist_ok=True)
df.to_csv('../data/processed/audited_2025.csv', index=False)
```
